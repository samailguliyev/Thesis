```python
import random
import pandas as pd

# Initialize DataFrame to hold the reshuffled DataFrame for each term
reshuffled_dfs = []

# Initialize DataFrame to hold comments that can't be assigned to any year for each term
# unassigned_comments = {}

for term in term_columns:
    # Subset and sort DataFrame for this term
    df_term = df_at_least_one_term[df_at_least_one_term[term] > 0].copy()
    df_term.sort_values(by=term, ascending=False, inplace=True)
    print(term, len(df_term))
    # Initialize Dictionary of Original Counts for this term
    original_counts = {}
    for year in range(2008, 2023):
        original_counts[year] = df_term[df_term['year'] == year][term].sum()

    # Initialize DataFrame to hold comments that can't be assigned to any year for this term
    # unassigned_comments_term = pd.DataFrame(columns=df_term.columns)

    for idx, row in df_term.iterrows():
        assigned = False
        for random_year in random.sample(range(2008, 2023), 15):  # Randomly shuffle the years to try
            # Check if this year assignment would cause the term count to go negative
            count = row[term]
            remaining_count = original_counts.get(random_year, 0) - count
            if remaining_count >= 0:
                # Assign the year to the comment
                df_term.loc[idx, 'reshuffled_year'] = random_year
                
                # Update the remaining counts in the dictionary
                original_counts[random_year] -= count

                assigned = True
                break  # Exit the for loop for this comment
        
    #     if not assigned:
    #         unassigned_comments_term = unassigned_comments_term.append(row)
    
    # unassigned_comments[term] = unassigned_comments_term.reset_index(drop=True)
    reshuffled_dfs.append(df_term)

# Concatenate all reshuffled dfs
final_df = pd.concat(reshuffled_dfs, ignore_index=True)


# Remove duplicate comments by texts (keep first)
# final_df.drop_duplicates(subset=['preprocessed_body'], keep='first', inplace=True)

# Reset the index of final_df
final_df.reset_index(drop=True, inplace=True)
```
