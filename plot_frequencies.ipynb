```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("default")


# Function to print number of comments per year
def print_comments_per_year(df, name):
    print(f"Total number of comments in {name}: {len(df)}")
    print(f"Number of comments per year in {name}:")
    print(df['reshuffled_year'].value_counts().sort_index())

# Print number of comments per year for each DataFrame
print_comments_per_year(df, "df")

# Plot individual term counts per year
for term in term_columns:
    plt.figure()
    df.groupby('reshuffled_year')[term].sum().plot(kind='bar')
    plt.title(f"{term} counts per year")
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.show()
    print(f"{term} counts per year:")
    print(df.groupby('reshuffled_year')[term].sum())  # Corrected this line
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("default")

# Assuming df is your DataFrame and term_columns contains your terms

# Normalize frequencies for each term
# normalized_df = df.copy()
for term in term_columns:
    # if term not in ["ring","hana","square","asana"]:
    df[term] = (df[term] - df[term].min()) / (df[term].max() - df[term].min())
    # df[term] = df[term]/100

    # print(term)

# Aggregate normalized frequencies per year
# The sum is automatically done across the columns (terms) for each group (year)
aggregated = (df.groupby('reshuffled_year')[term_columns].sum())/len(term_columns)

# Plot aggregated normalized frequencies per year
plt.figure()
aggregated.sum(axis=1).plot(kind='bar')  # Sum across all terms for each year
plt.title("Aggregated Normalized Frequencies per Year")
plt.xlabel('Year')
plt.ylabel('Normalized Frequency')
plt.show()
```
